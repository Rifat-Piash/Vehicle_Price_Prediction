```python
def findMissingValue(df):
    # Calculate the number of missing values per column
    missing_values = df.isnull().sum()

    # Calculate the percentage of missing values
    percent_missing = (missing_values / len(df)) * 100

    # Create a DataFrame with the results
    missing_data = pd.DataFrame({
        'Columns': df.columns,
        'Number_Of_MValue': missing_values,
        'Percentage_Of_MValue': percent_missing.round(2)
    })

    # Filter out columns with no missing data and sort
    missing_data = missing_data[missing_data['Number_Of_MValue'] > 0].sort_values('Number_Of_MValue', ascending=False).reset_index(drop=True)

    return missing_data
```

```python
# Finding out if all the columns that have missing values in it are float/int or object data type.

m_value = findMissingValue(df)

col = m_value['Columns']

data = pd.DataFrame(columns=['Column_Name', 'Data_Type'])
data_col = []
data_type = []

for i in col:
        
    data_col.append(i)
    data_type.append(df.select_dtypes(include=['float64', 'int64']).dtypes[1])
    
data['Column_Name']= data_col
data['Data_Type']= data_type
    
data
```

```python
# Removing data that contains less than 5% of missing value as we can drop 5% of observation if there is any missing value.

for i in df.columns:
    col_value = df[i].describe()[0]
    m_value = len(df) - col_value
    
    if m_value> 0.00:
        percntg_mvalue = m_value/len(df)
        if percntg_mvalue <= .05:
            
            df.dropna(axis=0,subset=[i],inplace=True)
```

```python
# Creating a function to execute the One-Hot Encoding 

def one_hot_top_x(df,variable,top_x_labels):
    
    for label in top_x_labels:
        df[variable+'_'+label] = np.where(df[variable]==label,1,0)
```

```python
# Implementing One-Hot Encoding to the selected variables

for i in variable_for_hot_encoding:
    
    if len(df[i].value_counts().sort_values(ascending=False).index)>20:
    
        top_x = [x for x in df[i].value_counts().sort_values(ascending=False).head(20).index]
    
    else:
        top_x = [x for x in df[i].value_counts().sort_values(ascending=False).index]
    
    one_hot_top_x(df,i,top_x)
    df.drop([i],axis=1,inplace=True)
```
